### List of unique members who have interacted at least once

```python
fhand = open('Exam.txt',encoding = 'utf8')
unique_members=[]
for line in fhand: 
    line=line.strip()  #removing spaces
    if "-" in line:    #getting only lines which has member
        line= line[line.find(":") +1:]  #getting only line which contain : once. so that we extract only chat lines further
       # print(line)
        if  ":" in line:
#                 print(line)
            line_chat = line[line.find("-")+1: line.find(":") ]  #extractuing only chat lines
           # print(line_chat)
            if line_chat not in unique_members: #finding if member is already present in unique_member list
                unique_members.append(line_chat) # appending unique member in list

print(unique_members) #printing list of unique members who have interacted at least once
```

### Extracting entire conversation happened on a particular date

#### if we want conversation with member

```python
def conversation_by_date(file, date):
    fhand = open(file,encoding = 'utf8')
    lst=[]
    for line in fhand: 
        line=line.strip() 
        #print(line)
        date_changed=date+","
        
        if line.startswith(date_changed):
            if "-" in line:    #getting only lines which has member
                line= line[line.find(":") +1:]  #getting only line which contain : once. so that we extract only chat lines further
               # print(line)
                if  ":" in line:
                     #print(line)
                    line_chat = line[line.find("-")+1: ]  #extractuing only chat lines
                    lst.append(line_chat)
    return(lst)
```

```python
conversation_by_date('Exam.txt', '2/2/22')
```

### Extract all the posts done by a particular person

```python
def conversation_by_person(file, person):
    fhand = open(file,encoding = 'utf8')
    lst_chat=[]
    modified_person = " " + person + ":"
    for line in fhand: 
        line=line.strip() 
        if "-" in line:    #getting only lines which has member
            line= line[line.find(":") +1:]  #getting only line which contain : once. so that we extract only chat lines further
            # print(line)
            if  ":" in line:
                #print(line)
                line_chat = line[line.find("-")+1: ]  #extractuing only chat lines
    #             print(line_chat)
                
                if line_chat.startswith(modified_person):
                    chat= line_chat[line_chat.find(":")+1:]
                    lst_chat.append(chat)
    return(lst_chat)
```

### Get the average length of messages posted by a particular person

#### if we are considering avg length by letters typed by person

```python
def avg_length_by_person(file, person):
    count=0
    lst_chat=[]
    fhand = open(file,encoding = 'utf8')
    
    modified_person = " " + person + ":"
    for line in fhand: 
        line=line.strip() 
        if "-" in line:    #getting only lines which has member
            line= line[line.find(":") +1:]  #getting only line which contain : once. so that we extract only chat lines further
            # print(line)
            if  ":" in line:
                #print(line)
                line_chat = line[line.find("-")+1: ]  #extractuing only chat lines
    #             print(line_chat)

                if line_chat.startswith(modified_person):
                    chat= line_chat[line_chat.find(":")+1:]
                    count+=1
                    trans = str.maketrans("",""," ")  #excluding spaces in messages
                    chat=chat.translate(trans) 
                    lst_chat.append(chat)

    letters_len= 0
    for message in lst_chat:
        letters_len= letters_len+len(message)
    avg_length=letters_len/ count
    return(avg_length)
```

```python
avg_length_by_person('Exam.txt', 'Gourab Nath Praxis')
```

#### if we are considering avg length by words typed by person

```python
def avg_length_by_person(file, person):
    count=0 #initiated count for counting no of  msg
    lst_chat=[] #initialized to add every msg by particular person in the list
    fhand = open(file,encoding = 'utf8') #opening file
    modified_person = " " + person + ":" #modifying person name to identify specific chat line for further use
    for line in fhand: #for each line in file
        line=line.strip() #cutting spaces
        if "-" in line:    #getting only lines which has member
            line= line[line.find(":") +1:]  #getting only line which contain : once. so that we extract only chat lines further
            # print(line)
            if  ":" in line: #if again colon is contained in line
                #print(line)
                line_chat = line[line.find("-")+1: ]  #extracting only chat lines
    #             print(line_chat)

                if line_chat.startswith(modified_person):
                    chat= line_chat[line_chat.find(":")+1:]
    #                 print(chat)
                    lst_chat.append(chat)

                    count+=1
                    lst_chat.append(chat)

    words_count= 0 #initializing words count
    # print(lst_chat)
    for message in lst_chat: # for each msg in chat of specific person
        message=message.split()  #splitiing msg into words
        words_count=words_count+len(message) #counting words in each msg
    avg_len_by_words=words_count/count #calculating avg length of msg by words
    return(avg_len_by_words) #returning avg length of msg
```

### Which part of the day (AM or PM) does this group experience most conversation?

```python
def time_most_conversation(file):
    fhand = open(file,encoding = 'utf8') #opening file
    am_conversation=0
    pm_conversation=0
    for line in fhand:
        line=line.strip() 
    #     print(line)
        if line.count(":") >= 2:  #checking if colon exists in string at least 2 times
            if line.find('-')==15: #reconfirming that we are extracting only chat line
                line_chat=line            
        #         print(line_chat)
                time= line[line.find(', ')+2:line.find(":")] #extracting exact time in hour
                time=int(time) #converting string type of time into int
                if time < 12:
                    am_conversation +=1
                if time >= 12:
                    pm_conversation +=1
    if am_conversation>pm_conversation:
        print("This group experience most conversation in AM part of the day")
    elif pm_conversation>am_conversation:
        print("This group experience most conversation in PM part of the day")
    else:
        print("This group experience equal conversation in PM and AM part of the day")            
```

```python
time_most_conversation('Exam.txt')
```

### frequency distribution of hours (considering only chats for frequency distribution)

```python
def hours_distribution(file):
    fhand = open(file,encoding = 'utf8') #opening file
    count_hours = {}
    for line in fhand:
        line=line.strip() 
        #print(line)
        if line.count(":") >= 2:  #checking if colon exists in string at least 2 times
            line_chat=line
    #         print(line_chat)
            time = line[line.find(", ")+2:line.find("-")] #extracting exact time in hour with minutes
    #         print(time)
            time_hour= time[:time.find(":")] #extracting time with hour only as its 24 hour clock we don't need minute
    #         print(time_hour)
    #         print(len(time_hour))
            if len(time_hour) <= 2:     #ensuring its only time and nothing else because chat can contain as well
                count_hours[time_hour]=count_hours.get(time_hour,0) +1
                keys = list(count_hours.keys())
                keys.sort()

    for key in keys:
        print(key, count_hours[key])  #printing frequency dtsrtibution of each hour
```
